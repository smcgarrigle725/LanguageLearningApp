## sklearn example:

Scikit Learn is one of the most popular machine learning libraries. It supports:

    - Classical ML algorithm  
    - Preprocessing tools  
    - Model evaluation  

```python
# Load dataset:
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df.describe()
```

```python
y = iris.target
y
```

```python
# DATA PRE-Porcessing: missing data, class imbalance, feature encoding, normalization/standardization, Feature Engineering, ...
# out of scope for this exercise
# 90% of ML is getting Data ready!
```

```python
# Split train/test data:
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.25, random_state=16)
# shuffle=True
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
```

```python
# Choose Model class
from sklearn.linear_model import LogisticRegression

# instantiate the model
model_reg = LogisticRegression(random_state=16)

# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html
```

```python
# Train Model: model.fit()
model_reg.fit(X_train, y_train)
```

```python
# Use Model to Predict: model.predict()
y_pred = model_reg.predict(X_test)
```

```python
# Evaluate Model: 
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_pred)
accuracy_score(y_test, y_pred)
```

```python
# plot confusion matrix: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.ConfusionMatrixDisplay.html
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=model_reg.classes_)
disp.plot()
plt.show()

```

# Artificial Neural Networks & Deep Learning:

Artificial Neural Networks (NN) are a class of algorithms consisting of layers of processing units known as artificial neurons. NNs are inspired by the human brain.
They come in different flavors and different architectures. Example:  

- A multilayer perceptron (MLP)
- Convolution Neural Networks (CNN)
- Recurrent Neural Networks (RNN)

<details><summary>More about Neural Networks</summary>  

1- Basic Components:
- Neurons (Nodes): Basic units that receive input, process it, and generate output
- Connections (Edges): Links between neurons that transmit signals
- Layers: Groups of neurons organized in a hierarchical structure:
    - Input Layer: Receives initial data
    - Hidden Layer(s): Processes information
    - Output Layer: Produces final results

2- How NNs Work:
- Each neuron receives inputs multiplied by weights
- The weighted sum goes through an activation function
- The result is passed to the next layer
- Mathematical representation for a single neuron
output = f(Sigma(wixi + b))

where:  
wi: are weights  
xi: are inputs  
b: is bias  
f: is the activation function  

3- Training Process:
- Forward propagation: Input data flows through the network
- Error calculation: Compare output with desired result
- Backpropagation: Adjust weights to minimize error
- Repeat until satisfactory performance is achieved


</details>

## Keras Example:


Keras is one of the most popular deep learning libraries. It is a high level neural network API that sits on top of more fundemental deep learning libraries. As of Keras 3.0, it supports the following backends:  
- Tensorflow
- PyTorch
- JAX

It has 2 [modes/classes](https://keras.io/api/):
- Sequential: supports linear stack of layers (easier, less felixble)
- Model: Functional API for complex architectures (more complex, more felxible)

It has several [layer types](https://keras.io/api/). Examples:
- Dense: Fully connected layer
- Conv2D: 2D Convolution layer
- MaxPooling2D: 2D Pooling layer
- Dropout: Dropout layer for regularization
- Flatten: Flattening layer
- LSTM: Long Short-Term Memory layer

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
from keras.utils import to_categorical
```

```python
## Normalize data
df = df/df.max(axis=0)
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.25, random_state=16)

## change y into one hot encoding:
y_train = to_categorical(y_train, num_classes=3)
y_test = to_categorical(y_test, num_classes=3)
```

```python
# build model

model = Sequential()
model.add(Dense(4, input_dim=4, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(3, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
# fit the keras model on the dataset
history = model.fit(X_train, y_train, epochs=300, batch_size=10, validation_split=0.2, shuffle=True)
```

```python
# Plot training curves
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.ylim(ymin=0)
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
# Plot loss curves
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.ylim(ymin=0)
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(X_test, y_test)
print(accuracy)
```
