# Activation Functions

Activation functions (AFs) are components in Neural Networks. They are important for several reasons:  

- Feature Learning
    - AFs make deep learning possible. Without activations, deep networks would collapse into a single linear layer
- Non-linearity
    - AFs allow networks to learn complex patterns
- Gradient Flow
    - AFs help control vanishing or exploding gradients
- Output Range Control
    - Different AFs have different output functions

- Enable Backpropagation
    - Derivatives of activation functions allow gradient-based learning



#### Without activation functions:
    - Networks would be limited to linear transformations
    - Couldn't learn complex patterns
    - Multiple layers would be equivalent to a single layer
    - Deep learning wouldn't work effectively


#### Common Activation Functions

- ReLU (Rectified Linear Unit) 
    - Formula: \[ f(x) = max(0, x) \]
    - Advantages:
        - Reduces vanishing gradient
        - Simple computation
        - Sparse activation
    - Use case: Hidden layers


- Sigmoid
    - Formula: \[ f(x) = \frac{1}{1 + e^{-x}} \]
    - Range: (0, 1)
    - Use cases:
        - Binary classification
        - Output gates in LSTM

- Tanh (Hyperbolic Tangent)
    - Formula: \[ f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} \]
    - Range: (-1, 1)
    - Use cases:
        - Hidden layers
        - RNN/LSTM cells

- Softmax
    - Formula: \[ f(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}} \]
    - Use case: Multi-class classification output


#### Plotting Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt


def plot_activation_functions():
    x = np.linspace(-5, 5, 100)
    
    plt.figure(figsize=(12, 8))
    plt.plot(x, np.maximum(0, x), label='ReLU')
    plt.plot(x, 1/(1 + np.exp(-x)), label='Sigmoid')
    plt.plot(x, np.tanh(x), label='Tanh')
    plt.plot(x, np.where(x > 0, x, 0.01*x), label='Leaky ReLU')
    
    plt.grid(True)
    plt.legend()
    plt.title('Activation Functions')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.ylim(-1.5,1.5)
    plt.show()

plot_activation_functions()
```

### How to choose an activation function?

- Sigmoid:
    - Use for binary classification
    - Outputs are between 0 and 1
    - Good for output layer in binary classification
    - Bad for hidden layers

- Tanh:
    - Use for hidden layers
    - Outputs are between -1 and 1
    - Good for hidden layers
    - Bad for output layer

- ReLU:
    - Use for hidden layers
    - Outputs are between 0 and infinity
    - Good for hidden layers
    - Bad for output layer

- Softmax:
    - Use for output layer in multi-class classification
    - Outputs are between 0 and 1
    - Bad for hidden layers
